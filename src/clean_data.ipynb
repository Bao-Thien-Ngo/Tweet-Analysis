```python
# Import the libraries
import pandas as pd
```

```python
# Read the CSV file
data = pd.read_csv('/content/Corona_NLP_test.csv')
```

```python
# Diplay the data
data
```

```python
# Show the list of column name in data
data.columns.tolist()
```

```python
# Find the null values in the table
data.isnull().sum()
```

```python
# Create a new dataframe data_2 contain 1 column OriginalTweet as a subset of the data dataframe
data_2 = data[['OriginalTweet']].copy() 
```

```python
# Diplay the data_2
data_2
```

```python
# Drop any NaN values have been dropped from the OriginalTweet column and change the data type to string
data_2['OriginalTweet'].dropna(inplace=True)
data_2['OriginalTweet'] = data_2['OriginalTweet'].astype(str)
```

```python
# Write the change data file to a new CSV file
data_2.to_csv("/content/tweet.csv", index=False)
```
